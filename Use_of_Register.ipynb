```python
import time
import random
import numpy as np
    
```

```python
# Data creation
N=1000000  #1e6
# Python List
X_list = [random.random() for _ in range(N)]
# NumPy Array
X_np= np.random.rand(N)

```

```python
# Implementation 1
start = time.time()
Y1 =[]
for i in range(len(X_list)-1):
    Y1.append(X_list[i+1] - X_list[i])

end = time.time()
print(f"Time taken for Python List: {end - start:.4f} seconds")
```

```python
# Implementation 2
start = time.time()
Y2 = []
x_now = X_list[0]
for i in range(1, len(X_list)):
    x_next = X_list[i]
    Y2.append(x_next - x_now)
    x_now = x_next
end = time.time()
print(f"Time taken for Python List (should be optimized): {end - start:.4f} seconds") # Python loop + Registers
```

```python
# Implementation 3
start = time.time()
Y3 = np.empty(N-1)
for i in range(N-1):
    Y3[i] = X_np[i+1] - X_np[i]
end = time.time()
print(f"Time taken for NumPy Array: {end - start:.4f} seconds")
```

```python
# Implementation 4
start = time.time()
Y4 = np.empty(N-1)
x_now = X_np[0]
for i in range(1, N):
    x_next = X_np[i]
    Y4[i-1] = x_next - x_now
    x_now = x_next
end = time.time()
print(f"Time taken for NumPy Array (should be optimized): {end - start:.4f} seconds")   # NumPy loop + Registers
```

```python
# Implementation 5
start = time.time()
Y5 = np.diff(X_np)
end = time.time()
print(f"Time taken for NumPy Array with np.diff: {end - start:.4f} seconds")  # Using NumPy's built-in function
```
